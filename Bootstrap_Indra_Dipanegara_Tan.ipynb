```python
#Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array([164, 167, 161, 170, 159, 155, 163, 172, 165, 180, 159, 158, 162, 170, 160, 173, 171, 161, 157, 175])
```

```python
n = len(data)
m= 500000
bootstrap_means = np.zeros(m)
for i in range (m):
  bootstrap_sample = np.random.choice(data, size = n, replace = True)
  bootstrap_means[i] = np.mean(bootstrap_sample)
mean_of_means = np.mean(bootstrap_means)
plt.hist(bootstrap_means, bins = 50, density = True, alpha = 0.5)
plt.xlabel("Mean")
plt.ylabel("Probability")
plt.title("Probability Distribution of the Mean")
plt.show()
confidence_interval = np.percentile(bootstrap_means, [5, 95])
print("Mean: ", mean_of_means)
print("90% Confidence Interval: ", confidence_interval)
```
